<h2>Confusion Matrix</h2>

```python
XY_color = [  (1,5 , 'Blue')  ,(1,6 ,'Blue')  , (2,4 ,'Blue')   , (2,5 ,'Red') ,
        (2,6 ,'Red')   , (2, 7 ,'Blue'), (3,-1 ,'Red')   , (3,1 ,'Red') ,
        (3, 5 ,'Blue') , (3,6 ,'Blue') , (3,7 ,'Blue')   , (4,1 ,'Blue') ,   (4,-1 ,'Red') ,
        (4,2 ,'Red')   , (5,1 ,'Red') , (5,2 ,'Red')    , (5,3 ,'Red')]
X = []
Y = []
Color = []
for x, y ,color in XY_color:

    X.append(x)
    Y.append(y)
    
    Color.append(color)
    
data = pd.DataFrame({ 'X':X , 'Y':Y ,'Color' : Color  })
```

```python
line_x = np.linspace(0.5, 5.3 ,10)
line_y = 0.75 * np.linspace(0.5 , 5.3 ,10)+1 
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize= (12, 8))
ax.scatter(data['X'], data['Y']  , alpha= 1 , s=200 , c= data['Color'])
ax.set_xlabel('X' ,fontsize=25)
ax.set_ylabel('Y' ,fontsize=25)
ax.plot(line_x, line_y )
ax.fill_between(line_x , line_y , 7.7 + 0 * line_y , alpha = 0.2  ,color='blue')
ax.fill_between(line_x , line_y , -1.4 + 0 * line_y , alpha = 0.2 ,color='red');
```
