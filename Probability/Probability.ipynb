    <h2 style="color: black;">Basic Concepts in Probability</h2>
    <h3 style="color: black;">1) Sample Space</h3>
    <p style="font-size: 16px; color: black;">
        The sample space, denoted by \( S \), is the set of all possible outcomes of a random experiment. It represents the entire universe of possible events that could occur.
    </p>
    <p style="font-size: 16px; color: black;">
        Mathematically, the sample space can be defined as:
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \( S = \{ s_1, s_2, \ldots, s_n \} \)
    </p>
    <p style="font-size: 16px; color: black;">
        where \( s_1, s_2, \ldots, s_n \) are the individual outcomes or elements of the sample space.
    </p>
    <h3 style="color: black;">2) Event</h3>
    <p style="font-size: 16px; color: black;">
        An event is a subset of the sample space, representing a particular outcome or combination of outcomes of interest. Events can be simple (single outcomes) or compound (multiple outcomes).
    </p>
    <p style="font-size: 16px; color: black;">
        Mathematically, an event \( A \) can be defined as:
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \( A \subseteq S \)
    </p>
    <h3 style="color: black;">3) Outcome</h3>
    <p style="font-size: 16px; color: black;">
        An outcome is a particular result or realization that occurs as a result of a random experiment. It is a specific element of the sample space.
    </p>
    <h3 style="color: black;">4) Probability Space</h3>
    <p style="font-size: 16px; color: black;">
        A probability space, denoted by \( (S, \mathcal{F}, P) \), consists of three components:
    </p>
    <ul style="font-size: 16px; color: black;">
        <li><b>Sample Space (\( S \)):</b> The set of all possible outcomes of the experiment.</li>
        <li><b>Event Space (\( \mathcal{F} \)):</b> A collection of subsets of the sample space, representing the events of interest.</li>
        <li><b>Probability Measure (\( P \)):</b> A function that assigns a probability to each event in the event space, satisfying certain axioms such as non-negativity, additivity, and total probability.</li>
    </ul>
</div>


<div style="border: 2px solid black; border-radius: 10px; padding: 15px;">
    <h2 style="color: black;">Probability Rules</h2>
    <h3 style="color: black;">1) Additive Rule</h3>
    <p style="font-size: 16px; color: black;">
        The additive rule, also known as the sum rule, states that the probability of the union of two events is equal to the sum of their individual probabilities minus the probability of their intersection.
    </p>
    <p style="font-size: 16px; color: black;">
        Mathematically, the additive rule can be expressed as:
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \( P(A \cup B) = P(A) + P(B) - P(A \cap B) \)
    </p>
    <h3 style="color: black;">2) Multiplicative Rule</h3>
    <p style="font-size: 16px; color: black;">
        The multiplicative rule, also known as the product rule, states that the probability of the intersection of two independent events is equal to the product of their individual probabilities.
    </p>
    <p style="font-size: 16px; color: black;">
        Mathematically, the multiplicative rule can be expressed as:
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \( P(A \cap B) = P(A) \times P(B) \)
    </p>
    <h3 style="color: black;">3) Complement Rule</h3>
    <p style="font-size: 16px; color: black;">
        The complement rule states that the probability of the complement of an event is equal to one minus the probability of the event itself.
    </p>
    <p style="font-size: 16px; color: black;">
        Mathematically, the complement rule can be expressed as:
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \( P(A') = 1 - P(A) \)
    </p>
</div>


<div style="border: 2px solid black; border-radius: 10px; padding: 15px;">
    <h2 style="color: black;">Types of Probability</h2>
    <h3 style="color: black;">1) Marginal Probability</h3>
    <p style="font-size: 16px; color: black;">
        Marginal probability refers to the probability of a single event occurring, disregarding any other events. It is calculated by summing or integrating the joint probabilities of all possible outcomes of the event of interest.
    </p>
    <p style="font-size: 16px; color: black;"><b>Mathematical Formula:</b></p>
    <p style="font-size: 16px; color: black;">
        If \( A \) is an event, \( P(A) \) represents the marginal probability of event \( A \).
    </p>
    <h3 style="color: black;">2) Conditional Probability</h3>
    <p style="font-size: 16px; color: black;">
        Conditional probability is the probability of one event occurring given that another event has already occurred. It is calculated by dividing the joint probability of both events by the probability of the condition event.
    </p>
    <p style="font-size: 16px; color: black;"><b>Mathematical Formula:</b></p>
    <p style="font-size: 16px; color: black;">
        The conditional probability of event \( A \) given event \( B \) is denoted by \( P(A|B) \) and is calculated as:
    </p>
    \[ P(A|B) = \frac{P(A \cap B)}{P(B)} \]
    <h3 style="color: black;">3) Joint Probability</h3>
    <p style="font-size: 16px; color: black;">
        Joint probability refers to the probability of two or more events occurring simultaneously. It is calculated by multiplying the probabilities of each individual event occurring.
    </p>
    <p style="font-size: 16px; color: black;"><b>Mathematical Formula:</b></p>
    <p style="font-size: 16px; color: black;">
        The joint probability of two events \( A \) and \( B \) is denoted by \( P(A \cap B) \) and is calculated as:
    </p>
    \[ P(A \cap B) = P(A) \times P(B|A) \]
</div>


<div style="border: 2px solid black; border-radius: 10px; padding: 15px;">
    <h2 style="color: black;">Probability Distributions</h2>
    <h3 style="color: black;">1) Discrete Probability Distributions</h3>
    <p style="font-size: 16px; color: black;">
        Discrete probability distributions represent the probabilities of outcomes associated with discrete random variables. These distributions are characterized by a finite or countably infinite number of possible outcomes, each with a distinct probability. In a discrete distribution, the probability mass function (PMF), denoted as \( P(X = x) \), assigns a probability to each possible value \( x \) of the random variable \( X \). The sum of all probabilities in a discrete distribution equals 1, expressed mathematically as:
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \[ \sum_{x} P(X = x) = 1 \]
    </p>
    <h3 style="color: black;">2) Continuous Probability Distributions</h3>
    <p style="font-size: 16px; color: black;">
        Continuous probability distributions describe the probabilities of outcomes associated with continuous random variables. Unlike discrete distributions, these distributions involve an uncountable number of possible outcomes, typically within a range or interval. In a continuous distribution, the probability density function (PDF), denoted as \( f(x) \), specifies the relative likelihood of observing different outcomes. The integral of the PDF over a range or interval yields the probability of the random variable falling within that range, expressed mathematically as:
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \[ P(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx \]
    </p>
    <h3 style="color: black;">3) Probability Mass Function (PMF)</h3>
    <p style="font-size: 16px; color: black;">
        The probability mass function (PMF) is used for discrete distributions and provides the probability of observing a specific value of the random variable. Mathematically, the PMF is defined as:
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \[ P(X = x) \]
    </p>
    <h3 style="color: black;">4) Probability Density Function (PDF)</h3>
    <p style="font-size: 16px; color: black;">
        The probability density function (PDF) is used for continuous distributions and describes the relative likelihood of observing different outcomes. The PDF is defined such that the area under the curve over any interval gives the probability of the random variable falling within that interval.
    </p>
    </p>
    <h3 style="color: black;">5) Cumulative Distribution Function (CDF)</h3>
        <p style="font-size: 16px; color: black;">
        The cumulative distribution function (CDF) for a discrete distribution gives the probability that the random variable \( X \) takes on a value less than or equal to a given value \( x \). It is defined as the sum of the probabilities of all outcomes less than or equal to \( x \), expressed mathematically as:
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \[ F(x) = P(X \leq x) = \sum_{x_i \leq x} P(X = x_i) \]
    </p>
    <p style="font-size: 16px; color: black;">
        The cumulative distribution function (CDF) for a continuous distribution gives the probability that the random variable \( X \) takes on a value less than or equal to a given value \( x \). It is defined as the integral of the PDF from negative infinity to \( x \), expressed mathematically as:
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \[ F(x) = P(X \leq x) = \int_{-\infty}^{x} f(t) \, dt \]
    </p>
    <p style="font-size: 16px; color: black;">
        Unlike a discrete distribution where probabilities are assigned to specific outcomes, in a continuous distribution, the probability of observing any single point is technically zero. Instead, probabilities are assigned to intervals or ranges of values.
    </p>


<div style="border: 2px solid black; border-radius: 10px; padding: 15px;">
    <h2 style="color: black;">Expected Value and Variance</h2>
    <h3 style="color: black;">1) Expected Value (Mean)</h3>
    <p style="font-size: 16px; color: black;">
        The expected value, also known as the mean, represents the average value of a random variable. It is a measure of the central tendency of a distribution and is calculated differently for discrete and continuous distributions.
    </p>
    <h4 style="color: black;">Discrete Distributions</h4>
    <p style="font-size: 16px; color: black;">
        For a discrete distribution, the expected value (\( E[X] \)) is computed as the weighted sum of all possible values of the random variable \( X \), each weighted by its corresponding probability. Mathematically, it is expressed as:
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \[ E[X] = \sum_{x} x \cdot P(X = x) \]
    </p>
    <h4 style="color: black;">Continuous Distributions</h4>
    <p style="font-size: 16px; color: black;">
        For a continuous distribution, the expected value (\( E[X] \)) is calculated by integrating the product of the variable \( x \) and the probability density function (\( f(x) \)) over the entire range of possible values. The integral represents the area under the curve, weighted by the likelihood of each outcome. Mathematically, it is expressed as:
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \[ E[X] = \int_{-\infty}^{\infty} x \cdot f(x) \, dx \]
    </p>
    <h3 style="color: black;">2) Variance and Standard Deviation</h3>
    <p style="font-size: 16px; color: black;">
        Variance measures the spread or dispersion of a distribution around its mean, while the standard deviation is the square root of the variance and provides a measure of the average distance between each data point and the mean. Similar to the expected value, variance and standard deviation are computed differently for discrete and continuous distributions.
    </p>
    <h4 style="color: black;">Discrete Distributions</h4>
    <p style="font-size: 16px; color: black;">
        For a discrete distribution, the variance (\( \text{Var}(X) \)) is calculated as the weighted sum of the squared deviations from the mean, each weighted by the corresponding probability. The standard deviation (\( \sigma \)) is the square root of the variance. Mathematically, they are expressed as:
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \[ \text{Var}(X) = \sum_{x} (x - E[X])^2 \cdot P(X = x) \]
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \[ \sigma = \sqrt{\text{Var}(X)} \]
    </p>
    <h4 style="color: black;">Continuous Distributions</h4>
    <p style="font-size: 16px; color: black;">
        For a continuous distribution, the variance (\( \text{Var}(X) \)) is computed by integrating the squared deviations from the mean weighted by the probability density function (\( f(x) \)) over the entire range of possible values. The standard deviation (\( \sigma \)) is the square root of the variance. Mathematically, they are expressed as:
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \[ \text{Var}(X) = \int_{-\infty}^{\infty} (x - E[X])^2 \cdot f(x) \, dx \]
    </p>
    <p style="font-size: 16px; color: black; text-align: center;">
        \[ \sigma = \sqrt{\text{Var}(X)} \]
    </p>
</div>

    <h2 style="color: black;">Common Probability Distributions</h2>
    <p style="font-size: 16px; color: black;">
        Probability distributions describe the likelihood of different outcomes in a random experiment or process. Here are some common probability distributions, along with their mathematical formulas, expected values, variances, and real-world applications:

    <h4 style="color: black;">Binomial Distribution</h4>

    </p>
    <ul style="font-size: 16px; color: black;">
        <li>Expected Value: \( E[X] = np \)</li>
        <li>Variance: \( \text{Var}(X) = np(1-p) \)</li>

    </ul>
    <h4 style="color: black;">Poisson Distribution</h4>
    <p style="font-size: 16px; color: black;">
        The Poisson distribution describes the number of events occurring in a fixed interval of time or space, given a constant rate of occurrence (\( \lambda \)).
    </p>
    <ul style="font-size: 16px; color: black;">
        <li>Expected Value: \( E[X] = \lambda \)</li>
        <li>Variance: \( \text{Var}(X) = \lambda \)</li>
        <li>Application: Modeling the number of phone calls received by a call center in a given hour, or the number of defects in a manufacturing process.</li>
    </ul>
    <h3 style="color: black;">2) Continuous Distributions</h3>
    <h4 style="color: black;">Normal Distribution</h4>
    <p style="font-size: 16px; color: black;">

    </p>
    <ul style="font-size: 16px; color: black;">
        <li>Expected Value: \( E[X] = \mu \)</li>
        <li>Variance: \( \text{Var}(X) = \sigma^2 \)</li>
        <li>Application: Modeling measurements such as height, weight, IQ scores, or errors in manufacturing processes.</li>
    </ul>
    <h4 style="color: black;">Exponential Distribution</h4>
    <p style="font-size: 16px; color: black;">
        The exponential distribution describes the time between events in a Poisson process, where events occur continuously and independently at a constant average rate (\( \lambda \)).
    </p>
    <ul style="font-size: 16px; color: black;">
        <li>Expected Value: \( E[X] = \frac{1}{\lambda} \)</li>
        <li>Variance: \( \text{Var}(X) = \frac{1}{\lambda^2} \)</li>
        <li>Application: Modeling the time between arrivals of customers at a service point, or the lifespan of electronic components.</li>
    </ul>

<div style="border: 2px solid black; border-radius: 10px; padding: 15px;">
    <h2 style="color: blue;">Binomial Distribution:</h2>
    <p style="font-size: 16px; color: black;">
        The binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials, where each trial has only two possible outcomes: success or failure. It is widely used in various fields, including statistics, biology, and finance.
    </p>
    <h3 style="color: black;">Probability Mass Function (PMF)</h3>
    <p style="font-size: 16px; color: black;">
        The probability mass function (PMF) of the binomial distribution gives the probability of obtaining exactly \( k \) successes in \( n \) trials, where \( p \) is the probability of success in each trial.
    </p>
    <p style="font-size: 16px; color: black;">
        The PMF of the binomial distribution is given by the formula:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ P(X = k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k} \]
    </p>
    <h3 style="color: black;">Cumulative Distribution Function (CDF)</h3>
    <p style="font-size: 16px; color: black;">
        The cumulative distribution function (CDF) of the binomial distribution gives the probability that the number of successes is less than or equal to a certain value \( k \).
    </p>
    <p style="font-size: 16px; color: black;">
        The CDF of the binomial distribution can be calculated by summing the probabilities of all possible outcomes up to \( k \):
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ F(k; n, p) = \sum_{i=0}^{k} \binom{n}{i} \cdot p^i \cdot (1 - p)^{n - i} \]
    </p>
    <h3 style="color: black;">Mean and Variance</h3>
    <p style="font-size: 16px; color: black;">
        The mean (\( \mu \)) and variance (\( \sigma^2 \)) of the binomial distribution are given by:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ \mu = np \]
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ \sigma^2 = np(1 - p) \]
    </p>
    <p style="font-size: 16px; color: black;">
        These parameters represent the expected number of successes and the spread or variability of the distribution, respectively.
    </p>
</div>


<div style="border: 2px solid black; border-radius: 10px; padding: 15px;">
    <h2 style="color: blue;">Poisson Distribution:</h2>
    <p style="font-size: 16px; color: black;">
        The Poisson distribution is a discrete probability distribution that describes the number of events occurring in a fixed interval of time or space, given that these events occur with a known constant mean rate and are independent of the time since the last event. It is commonly used to model the number of arrivals in a fixed period of time, such as the number of customers arriving at a service point in a given hour, the number of phone calls received by a call center in a minute, or the number of particles emitted by a radioactive source in a given time interval.
    </p>
    <h3 style="color: black;">Probability Mass Function (PMF)</h3>
    <p style="font-size: 16px; color: black;">
        The probability mass function (PMF) of the Poisson distribution gives the probability of observing exactly \( k \) events in a fixed interval of time or space, given that the average rate of occurrence of events is \( \lambda \) per interval.
    </p>
    <p style="font-size: 16px; color: black;">
        The PMF of the Poisson distribution is given by the formula:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ P(X = k) = \frac{{e^{-\lambda} \lambda^k}}{{k!}} \]
    </p>
    <h3 style="color: black;">Cumulative Distribution Function (CDF)</h3>
    <p style="font-size: 16px; color: black;">
        The cumulative distribution function (CDF) of the Poisson distribution gives the probability that the number of events is less than or equal to a certain value \( k \).
    </p>
    <p style="font-size: 16px; color: black;">
        The CDF of the Poisson distribution can be calculated by summing the probabilities of all possible outcomes up to \( k \):
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ F(k; \lambda) = \sum_{i=0}^{k} \frac{{e^{-\lambda} \lambda^i}}{{i!}} \]
    </p>
    <h3 style="color: black;">Mean and Variance</h3>
    <p style="font-size: 16px; color: black;">
        The mean (\( \mu \)) and variance (\( \sigma^2 \)) of the Poisson distribution are both equal to the average rate of occurrence of events, denoted by \( \lambda \).
    </p>
    <p style="font-size: 16px; color: black;">
        These parameters represent the expected number of events in the given interval and the spread or variability of the distribution, respectively.
    </p>
</div>


<div style="border: 2px solid black; border-radius: 10px; padding: 15px;">
    <h2 style="color: blue;">Geometric Distribution:</h2>
    <p style="font-size: 16px; color: black;">
        The geometric distribution is a discrete probability distribution that describes the number of trials needed to achieve the first success in a sequence of independent Bernoulli trials, where each trial has only two possible outcomes: success or failure. It is often used to model the number of failures before the first success in repeated trials of an experiment with a constant probability of success \( p \).
    </p>
    <h3 style="color: black;">Probability Mass Function (PMF)</h3>
    <p style="font-size: 16px; color: black;">
        The probability mass function (PMF) of the geometric distribution gives the probability of needing exactly \( k \) trials to achieve the first success.
    </p>
    <p style="font-size: 16px; color: black;">
        The PMF of the geometric distribution is given by the formula:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ P(X = k) = (1 - p)^{k - 1} \cdot p \]
    </p>
    <h3 style="color: black;">Cumulative Distribution Function (CDF)</h3>
    <p style="font-size: 16px; color: black;">
        The cumulative distribution function (CDF) of the geometric distribution gives the probability that the number of trials needed to achieve the first success is less than or equal to a certain value \( k \).
    </p>
    <p style="font-size: 16px; color: black;">
        The CDF of the geometric distribution can be calculated by summing the probabilities of all possible outcomes up to \( k \):
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ F(k; p) = 1 - (1 - p)^k \]
    </p>
    <h3 style="color: black;">Mean and Variance</h3>
    <p style="font-size: 16px; color: black;">
        The mean (\( \mu \)) and variance (\( \sigma^2 \)) of the geometric distribution are given by:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ \mu = \frac{1}{p} \]
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ \sigma^2 = \frac{1 - p}{p^2} \]
    </p>
    <p style="font-size: 16px; color: black;">
        These parameters represent the expected number of trials needed to achieve the first success and the spread or variability of the distribution, respectively.
    </p>
</div>


<div style="border: 2px solid black; border-radius: 10px; padding: 15px;">
    <h2 style="color: green;">Normal (Gaussian) Distribution:</h2>
    <p style="font-size: 16px; color: black;">
        The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetric about its mean, with the probability density function (PDF) forming the well-known bell-shaped curve. It is widely used in various fields, including statistics, physics, social sciences, and finance, due to its mathematical tractability and its occurrence in natural phenomena.
    </p>
    <h3 style="color: black;">Probability Density Function (PDF)</h3>
    <p style="font-size: 16px; color: black;">
        The probability density function (PDF) of the normal distribution gives the probability density at a given point \( x \) along the real number line.
    </p>
    <p style="font-size: 16px; color: black;">
        The PDF of the normal distribution is given by the formula:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ f(x; \mu, \sigma) = \frac{1}{{\sigma \sqrt{2\pi}}} e^{-\frac{(x - \mu)^2}{2\sigma^2}} \]
    </p>
    <h3 style="color: black;">Cumulative Distribution Function (CDF)</h3>
    <p style="font-size: 16px; color: black;">
        The cumulative distribution function (CDF) of the normal distribution gives the probability that a random variable \( X \) is less than or equal to a given value \( x \).
    </p>
    <p style="font-size: 16px; color: black;">
        The CDF of the normal distribution can be calculated using tables or computational methods and does not have a simple closed-form expression.
    </p>
    <h3 style="color: black;">Mean and Variance</h3>
    <p style="font-size: 16px; color: black;">
        The mean (\( \mu \)) and variance (\( \sigma^2 \)) of the normal distribution determine its location and spread, respectively.
    </p>
    <p style="font-size: 16px; color: black;">
        The mean (\( \mu \)) of the normal distribution is given by the integral:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ \mu = \int_{-\infty}^{\infty} x \cdot f(x; \mu, \sigma) \, dx \]
    </p>
    <p style="font-size: 16px; color: black;">
        The variance (\( \sigma^2 \)) of the normal distribution is given by the integral:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ \sigma^2 = \int_{-\infty}^{\infty} (x - \mu)^2 \cdot f(x; \mu, \sigma) \, dx \]
    </p>
    <p style="font-size: 16px; color: black;">
        These integrals represent the expected value of \( x \) and the spread or dispersion around the mean, respectively.
    </p>
    <img src="https://upload.wikimedia.org/wikipedia/commons/thumb/7/74/Normal_Distribution_PDF.svg/720px-Normal_Distribution_PDF.svg.png" alt="Uniform Distribution Graph" style="display: block; margin-left: auto; margin-right: auto; max-width: 80%;">
</div>


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the seed for reproducibility
np.random.seed(42)

# Define the parameters for the normal distributions
params = [
    (0, np.sqrt(0.2)),
    (0, np.sqrt(1.0)),
    (0, np.sqrt(5.0)),
    (-2, np.sqrt(0.5))
]

# Generate datasets
datasets = [np.random.normal(mu, sigma, 100000) for mu, sigma in params]

# Plot density functions
plt.figure(figsize=(7, 5))
for data, (mu, sigma) , color in zip(datasets, params , ['darkblue' , 'red' , 'orange' , 'green']):
    label = f'$\\mu = {mu}$, $\\sigma^2 = {round(sigma**2 , 2 )}$'
    sns.kdeplot(data, label=label , linewidth = 2 , color = color)

plt.title('Density Plots of Normal Distributions')
plt.ylabel('$\\phi_{\\mu, \\sigma^2}(X)$')
plt.xlabel('$X$')
plt.legend()
plt.xlim(-5 , 5)
plt.grid()
plt.show()
```

<div style="border: 2px solid black; border-radius: 10px; padding: 15px;">
    <h2 style="color: green;">Exponential Distribution:</h2>
    <p style="font-size: 16px; color: black;">
        The exponential distribution is a continuous probability distribution that describes the time between events in a Poisson process, where events occur continuously and independently at a constant average rate \( \lambda \). It is commonly used to model the time until the next event occurs, such as the time until a radioactive particle decays, the time until a customer arrives at a service point, or the time until a component fails.
    </p>
    <h3 style="color: black;">Probability Density Function (PDF)</h3>
    <p style="font-size: 16px; color: black;">
        The probability density function (PDF) of the exponential distribution gives the probability density at a given time \( x \) along the real number line.
    </p>
    <p style="font-size: 16px; color: black;">
        The PDF of the exponential distribution is given by the formula:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ f(x; \lambda) = \lambda e^{-\lambda x} \]
    </p>
    <h3 style="color: black;">Cumulative Distribution Function (CDF)</h3>
    <p style="font-size: 16px; color: black;">
        The cumulative distribution function (CDF) of the exponential distribution gives the probability that a random variable \( X \) is less than or equal to a given value \( x \).
    </p>
    <p style="font-size: 16px; color: black;">
        The CDF of the exponential distribution is given by the formula:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ F(x; \lambda) = 1 - e^{-\lambda x} \]
    </p>
    <h3 style="color: black;">Mean and Variance</h3>
    <p style="font-size: 16px; color: black;">
        The mean (\( \mu \)) and variance (\( \sigma^2 \)) of the exponential distribution are both equal to the inverse of the rate parameter \( \lambda \).
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ \mu = \frac{1}{\lambda} \]
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ \sigma^2 = \frac{1}{\lambda^2} \]
    </p>
    <p style="font-size: 16px; color: black;">
        These parameters represent the average time until an event occurs and the spread or variability of the distribution, respectively.
    </p>
    <img src="https://upload.wikimedia.org/wikipedia/commons/thumb/f/f5/Exponential_distribution_pdf_-_public_domain.svg/1280px-Exponential_distribution_pdf_-_public_domain.svg.png" alt="Uniform Distribution Graph" style="display: block; margin-left: auto; margin-right: auto; max-width: 80%;">
</div>


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the seed for reproducibility
np.random.seed(42)

# Generate datasets
lambda_values = [0.5, 1, 1.5]
datasets = [np.random.exponential(1/lambda_val, 100000) for lambda_val in lambda_values]

# Plot density functions
plt.figure(figsize=(6, 5))
for data, lambda_val , color in zip(datasets, lambda_values , ['red' , 'green' , 'cyan']):
    sns.kdeplot(data, label=f'$\\lambda = {lambda_val}$' , color = color)

plt.title('Density Plots of Exponential Distributions')
plt.xlabel('$X$')
plt.ylabel('$P(X)$')
plt.grid()
plt.legend()
plt.xlim(0 , 5)
plt.show()
```

<div style="border: 2px solid black; border-radius: 10px; padding: 15px;">
    <h2 style="color: black;">Uniform Distribution:</h2>
    <p style="font-size: 16px; color: black;">
        The uniform distribution is a continuous probability distribution where every outcome within a given range is equally likely to occur. It is often used when no particular outcome is expected to be more likely than another within the specified range.
    </p>
    <h3 style="color: black;">Probability Density Function (PDF)</h3>
    <p style="font-size: 16px; color: black;">
        The probability density function (PDF) of the uniform distribution gives the probability density at a given point \( x \) along the real number line.
    </p>
    <p style="font-size: 16px; color: black;">
        The PDF of the uniform distribution is constant within the specified range and zero outside that range:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ f(x; a, b) = \frac{1}{b - a} \text{ for } a \leq x \leq b \]
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ f(x; a, b) = 0 \text{ for } x < a \text{ or } x > b \]
    </p>
    <h3 style="color: black;">Cumulative Distribution Function (CDF)</h3>
    <p style="font-size: 16px; color: black;">
        The cumulative distribution function (CDF) of the uniform distribution gives the probability that a random variable \( X \) is less than or equal to a given value \( x \).
    </p>
    <p style="font-size: 16px; color: black;">
        The CDF of the uniform distribution is given by the formula:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ F(x; a, b) = \frac{x - a}{b - a} \text{ for } a \leq x \leq b \]
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ F(x; a, b) = 0 \text{ for } x < a \]
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ F(x; a, b) = 1 \text{ for } x \geq b \]
    </p>
    <h3 style="color: black;">Mean and Variance</h3>
    <p style="font-size: 16px; color: black;">
        The mean (\( \mu \)) and variance (\( \sigma^2 \)) of the uniform distribution are given by:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ \mu = \frac{a + b}{2} \]
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ \sigma^2 = \frac{(b - a)^2}{12} \]
    </p>
    <p style="font-size: 16px; color: black;">
        These parameters represent the midpoint of the range and the spread or variability of the distribution, respectively.
    </p>

<div style="border: 2px solid black; border-radius: 10px; padding: 15px;">
    <h2 style="color: black;">Beta Distribution:</h2>
    <p style="font-size: 16px; color: black;">
        The beta distribution is a continuous probability distribution defined on the interval [0, 1]. It is commonly used to model random variables that represent proportions or probabilities, such as the success rate of a binomial process or the probability of an event occurring within a certain timeframe. The shape of the beta distribution can be highly flexible, making it suitable for a wide range of applications.
    </p>
    <h3 style="color: black;">Probability Density Function (PDF)</h3>
    <p style="font-size: 16px; color: black;">
        The probability density function (PDF) of the beta distribution gives the probability density at a given point \( x \) along the interval [0, 1].
    </p>
    <p style="font-size: 16px; color: black;">
        The PDF of the beta distribution is given by the formula:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ f(x; \alpha, \beta) = \frac{x^{\alpha - 1} (1 - x)^{\beta - 1}}{B(\alpha, \beta)} \]
    </p>
    <p style="font-size: 16px; color: black;">
        where \( B(\alpha, \beta) \) is the beta function.
    </p>
    <h3 style="color: black;">Cumulative Distribution Function (CDF)</h3>
    <p style="font-size: 16px; color: black;">
        The cumulative distribution function (CDF) of the beta distribution does not have a simple closed-form expression and is typically computed numerically.
    </p>
    <h3 style="color: black;">Mean and Variance</h3>
    <p style="font-size: 16px; color: black;">
        The mean (\( \mu \)) and variance (\( \sigma^2 \)) of the beta distribution depend on the shape parameters \( \alpha \) and \( \beta \).
    </p>
    <p style="font-size: 16px; color: black;">
        The mean and variance are given by the formulas:
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ \mu = \frac{\alpha}{\alpha + \beta} \]
    </p>
    <p style="font-size: 16px; color: black; margin-left: 40px;">
        \[ \sigma^2 = \frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)} \]
    </p>
    <p style="font-size: 16px; color: black;">
        These parameters determine the location and spread of the distribution, respectively.
    </p>
    <img src="https://upload.wikimedia.org/wikipedia/commons/thumb/f/f3/Beta_distribution_pdf.svg/531px-Beta_distribution_pdf.svg.png" alt="Uniform Distribution Graph" style="display: block; margin-left: auto; margin-right: auto; max-width: 80%;">
</div>


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the seed for reproducibility
np.random.seed(42)

# Define the parameters for the beta distributions
params = [
    (0.5, 0.5),
    (5, 1),
    (1, 3),
    (2, 2),
    (2, 5)
]

list_colors = ['red' , 'blue' , 'green' , 'purple' , 'orange']

# Generate datasets
datasets = [np.random.beta(alpha, beta, 100000) for alpha, beta in params]

# Plot density functions
plt.figure(figsize=(7, 5))
for data, (alpha, beta) , color in zip(datasets, params , list_colors ):
    label = f'$\\alpha = {alpha}$, $\\beta = {beta}$'
    sns.kdeplot(data, label=label , color = color )

plt.title('Density Plots of Beta Distributions')
plt.xlabel('$X$')
plt.ylabel('$\\phi_{\\alpha, \\beta}(X)$')
plt.legend()
plt.grid()
plt.show()
```
